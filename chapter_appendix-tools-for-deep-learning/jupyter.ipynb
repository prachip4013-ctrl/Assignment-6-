```python

# Membership Operators
a = [10, 20, 30, 40]

print("List:", a)
print(20 in a)
print(50 not in a)

# List Concatenation, Repetition, and Functions
a = [1, 2, 3]
b = [4, 5]

c = a + b
print(c)

print(a * 2)

print(len(a))
print(max(a))
print(min(a))

# Accessing List Elements
a = [10, 20, 30, 40, 50]

print(a[0])      # First element
print(a[-1])     # Last element
print(a[1:4])    # Slicing

# Modifying a List
a = [1, 2, 3, 4]

a[1] = 20
print(a)

a.append(5)
print(a)

a.insert(1, 100)
print(a)

a.remove(20)
print(a)
```
